```python
class Book:
    def __init__(self, title, isbn, price):
        self.title = title
        self.isbn = isbn
        self.price = price


b1 = Book("data","1443",30.0)
b2 = Book("learn","1445",40.0)
b3 = Book("analysis","1446",50.0)

def total_price(books):
  s = 0
  for b in books:
    s += b.price
  return s

print(total_price([b1,b2,b3]))
```

```python
class Book:
    def __init__(self, title, price, keywords):
        self.title = title; self.price = price; self.keywords = set(keywords)
    def __lt__(self, rhs):
        if self.price != rhs.price: return self.price < rhs.price
        else: return self.title < rhs.title
    def __str__(self):
        return self.title + ' ($' + str(self.price) + ')'
    def update_price(self, new_price):
        self.price = new_price
    def get_common_keywords(self, other):
        return self.keywords & other.keywords
b1 = Book('Python',    99, ['code','computer'])
b2 = Book('Calculus', 199, ['maths'])
b3 = Book('Physics',   99, ['science','maths'])
b1.update_price(199)
print(Book.get_common_keywords(b2,b3))
if b3 < b2: print(b1)
books = [b1,b2,b3]
books.sort()
print(books[0],',',books[1],',',books[2])
# 'Physics ($99) , Calculus ($199) , Python ($199)'
```

```python
class Rational:
  def gcd(a,b):
    while b != 0:
      a,b = b, a%b
    return a

  def __init__(self, n, d):
    g = Rational.gcd(n,d)
    self.n = n//g     # numerator
    self.d = d//g      # denominator

  def add(self,x):
    d = self.d * x.d
    n = self.n * x.d + x.n * self.d
    return Rational(n,d)

  def mul(self,x):
    d = self.d * x.d
    n = self.n * x.n
    return Rational(n,d)

  def to_str(self):
    return "'"+ str(self.n) +"/" + str(self.d) +"'"
```

```python
a = Rational(1,2)
b = Rational(3,4)
print(a)
print(a.to_str())
c = a.add(b)
print(c.to_str())
```

```python
class Rational:
  def gcd(a,b):
    while b != 0:
      a,b = b, a%b
    return a

  def __init__(self, n, d):
    g = Rational.gcd(n,d)
    self.n = n//g     # numerator
    self.d = d//g      # denominator

  def __add__(self,x):
    d = self.d * x.d
    n = self.n * x.d + x.n * self.d
    return Rational(n,d)

  def __mul__(self,x):
    d = self.d * x.d
    n = self.n * x.n
    return Rational(n,d)

  def __str__(self):
    return "<"+ str(self.n) +"/" + str(self.d) +">"
```

คลาสของไพ่

ข้างล่างนี้แสดงการเรียกใช้คลาส Card ซึ่งแทนไพ่ 1 ใบ ประกอบด้วย ค่าของไพ่ (value) ซึ่งเป็นสตริง "A", "2", "3", … ,
"10", "J", "Q", "K" และ ดอกของไพ่ (suit) ซึ่งเป็นสตริงเช่นกัน "club", "diamond", "heart", "spade"
โปรแกรมข้างล่างนี้รับไพ่เข้ามาหลายใบมาสร้างเป็นลิสต์ของไพ่ (cards) และมีการเรียกใช้เมท็อดต่าง ๆ ของคลาส Card ให้นักศึกษา
เขียนเมท็อดต่าง ๆ ของคลาส Card ให้สมบูรณ์ (ห้ามแก้ไขบริเวณที่มีพื้นหลังสีเทา)

Screenshot 2567-09-16 at 09.43.21.png

รายละเอียดต่าง ๆ ของคลาส Card และเมท็อดของคลาส Card

• เมท็อด getScore จะคืนค่าคะแนนของไพ่เป็นจานวนเต็ม ตามกฎดังนี้

o ไพ่ที่มีค่า A จะมีคะแนน 1 คะแนน

o ไพ่ที่มีค่า 2 ถึง 10 จะมีคะแนนเท่ากับค่าของไพ่ คือ 2 ถึง 10 คะแนน ตามลำดับ

o ไพ่ที่มีค่า J, Q, K จะมีคะแนน 10 คะแนน

• เมท็อด sum จะคืนค่าผลรวมคะแนนของไพ่สองใบและ mod ด้วย 10 เช่น

o Card.sum(Card("7", "club"), Card("2", "heart")) ได้ผลลัพธ์เป็น 9

o Card.sum(Card("J", "spade"), Card("5", "diamond")) ได้ผลลัพธ์เป็น 5

• การเรียงลำดับของไพ่เป็นดังนี้

o ค่าของไพ่เรียงตามลำดับดังนี้ 3 < 4 < 5 < … < 10 < J < Q < K < A < 2

o ดอกของไพ่เรียงตามลำดับดังนี้ club < diamond < heart < spade

o ถ้าไพ่สองใบมีค่าไม่เท่ากัน ไพ่ที่มีค่ามากกว่าจะเป็นไพ่ที่มากกว่า

o ถ้าไพ่สองใบมีค่าเท่ากัน ไพ่ที่มีดอกสูงกว่าจะเป็นไพ่ที่มากกว่า

```python
class Card:
    value_order = {'3': 1, '4': 2, '5': 3, '6': 4, '7': 5, '8': 6, '9': 7, '10': 8, 'J': 9, 'Q': 10, 'K': 11, 'A': 12, '2': 13}
    suit_order = {'club': 1, 'diamond': 2, 'heart': 3, 'spade': 4}

    def __init__(self, value, suit):
        self.value = value
        self.suit = suit

    def __str__(self):
        return f"{self.value} {self.suit}"

    def getScore(self):
        if self.value == "A":
            return 1
        elif self.value in ["J", "Q", "K"]:
            return 10
        else:
            return int(self.value)

    def sum(self, other):
        return (self.getScore() + other.getScore()) % 10

    def __lt__(self, other):
        # เปรียบเทียบตามค่าไพ่ก่อน ถ้าเท่ากันให้เปรียบเทียบตามดอกไพ่
        if Card.value_order[self.value] != Card.value_order[other.value]:
            return Card.value_order[self.value] < Card.value_order[other.value]
        else:
            return Card.suit_order[self.suit] < Card.suit_order[other.suit]

# การรับข้อมูลและการทำงาน
n = int(input("กรุณากรอกจำนวนไพ่: "))
cards = []
for i in range(n):
    value, suit = input("กรอกค่าและดอกของไพ่: ").split()
    cards.append(Card(value, suit))

# แสดงคะแนนของไพ่แต่ละใบ
for i in range(n):
    print(cards[i].getScore())

print("---------------------------")

# แสดงผลรวมคะแนนของไพ่สองใบที่อยู่ติดกัน
for i in range(n - 1):
    print(cards[i].sum(cards[i + 1]))

print("---------------------------")

# เรียงลำดับไพ่และแสดงผล
cards.sort()
for i in range(n):
    print(cards[i])
```
